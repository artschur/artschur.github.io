```python
import pandas as pd


df = pd.read_csv('winemag-data_first150k.csv')
```

```python


df.drop("winery",axis=1)
```

```python

# Assuming df is your DataFrame
filtered_df = df.query('points > 94 and country == "US"')
sorted_df = filtered_df.sort_values(by=["price"])

nndf = sorted_df.dropna()
ndf = nndf.drop('Unnamed: 0', axis=1 )
nnewdf = ndf.reset_index()
newdf = nnewdf.drop('index', axis=1)
newdf.head(15)

newdf.to_csv('newdf')

# Now you can work with the sorted and filtered DataFrame
```
