```python
import pandas as pd
import numpy as np
import string as str
df = pd.read_excel('Cópia de Lista de Vendas CS Brasil - 1158 - 21.02.2024.xlsx', header=None)
pd.set_option('display.max_rows', None)  # Display all rows
pd.set_option('display.max_columns', None)  # Display all columns
pd.set_option('display.precision', 2)  # Set float precision
```

```python
df = df.dropna()
new_header = df.iloc[0] #grab the first row for the header
df = df[1:] #take the data less the header row
df.columns = new_header #set the header row as the df header

ndf = df

ndf.head()
```

```python
ndf = ndf.drop(['EMPRESA', 'PLACA', 'ANO FAB',
                'ENDEREÇOS', 'DESCONTO/FIPE', 
                'LINK VISTORIA', 'AÇÃO CARTÃO'],
                axis=1)
```

```python
ndf['difference'] = ndf['FIPE'] - ndf['VALOR MÍNIMO']

ndf.head()
```

```python
difference_column = ndf['difference']
ndf = pd.concat([difference_column, ndf.drop(columns=['difference'])], axis=1)
```

```python
ndf = ndf.sort_values('difference', ascending=False)

```

```python
ndf = ndf[ndf['PÁTIO'].str.startswith('SP')]

ndf.head(15)
```

```python
fdf = ndf[~ndf['MODELO'].str.startswith('HILUX')]

fdf['KM'] = pd.to_numeric(fdf['KM'], errors='coerce')

# Remove rows with NaN values (which are the rows containing strings)
fdf = fdf.dropna(subset=['KM'])
```

```python

# Filter out rows with 'KM' values less than 50000
fdf = fdf[fdf['KM'] <= 50000]

# At this point, 'KM' is still a column in fdf, so you should not convert it again to numeric
fdf = fdf.sort_values('KM')
# Display the first few rows
fdf.head(15)
```

```python
fdf['MODELO'].value_counts()
gol =  fdf.loc[fdf['MODELO'] == 'GOL MSI 1.6 MEC']
gol = gol.sort_values('difference', ascending=False)

gol.head(10)

```
